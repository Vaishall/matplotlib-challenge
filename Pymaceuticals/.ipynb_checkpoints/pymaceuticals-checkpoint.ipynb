```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = os.path.join('data','mouse_drug_data.csv')
clinical_trial_data_to_load = os.path.join('data','clinicaltrial_data.csv')

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data=pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
total_data=pd.merge(clinical_trial_data, mouse_drug_data, how='left')

# Display the data table for preview
total_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_timepoint_group=total_data.groupby(['Drug','Timepoint'])
# Convert to DataFrame
mean_volume_data=pd.DataFrame(drug_timepoint_group['Tumor Volume (mm3)'].mean())
# Preview DataFrame
mean_volume_data.head()
```

```python
# Convert to DataFrame
sem_volume_data=pd.DataFrame(drug_timepoint_group['Tumor Volume (mm3)'].sem())
# Preview DataFrame
#sem_volume_data
sem_volume_data.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_volume_expanded=mean_volume_data.unstack('Drug')
mean_volume_expanded.columns=mean_volume_expanded.columns.droplevel(0)
sem_volume_expanded=sem_volume_data.unstack('Drug')
sem_volume_expanded.columns=sem_volume_expanded.columns.droplevel(0)
# Preview that Reformatting worked
mean_volume_expanded.head()
```

```python
# Generate the Plot (with Error Bars)
timepoints=list(mean_volume_expanded.index)

fig, ax =plt.subplots()
ax.set_title('Tumor Response to Treatment')
#get the data for each drug treatment (with error bars)
ax.errorbar(timepoints,mean_volume_expanded['Capomulin'], yerr=sem_volume_expanded['Capomulin'],
            fmt='--',color='r', marker='o', label='Capomulin')
ax.errorbar(timepoints,mean_volume_expanded['Infubinol'], yerr=sem_volume_expanded['Infubinol'],
            fmt='--', color='b', marker='^', label='Infubinol')
ax.errorbar(timepoints,mean_volume_expanded['Ketapril'], yerr=sem_volume_expanded['Ketapril'],
            fmt='--', color='g', marker='s', label='Ketapril')
ax.errorbar(timepoints,mean_volume_expanded['Placebo'], yerr=sem_volume_expanded['Placebo'],
            fmt='--', color='black', marker='d', label='Placebo')
#format our plot
ax.set_xlim(min(timepoints)-2, max(timepoints)+2)
ax.set_ylim(min(min(mean_volume_expanded['Capomulin']),
               min(mean_volume_expanded['Infubinol']),
               min(mean_volume_expanded['Ketapril']),
               min(mean_volume_expanded['Placebo']))-5,
           max(max(mean_volume_expanded['Capomulin']),
              max(mean_volume_expanded['Infubinol']),
              max(mean_volume_expanded['Ketapril']),
              max(mean_volume_expanded['Placebo']))+5)
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Tumor Volume (mm3)')
ax.legend(loc='best')
ax.grid()
# Save the Figure
plt.savefig('Images/treatment.png')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Convert to DataFrame
mice_count_data=pd.DataFrame(drug_timepoint_group['Mouse ID'].count())
# Preview DataFrame
mice_count_data.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_expanded=mice_count_data.unstack('Drug')
mice_count_expanded.columns=mice_count_expanded.columns.droplevel(0)
# Preview the Data Frame
mice_count_expanded.head()
```

```python
# Generate the Plot (Accounting for percentages)
timepoints=list(mice_count_expanded.index)

fig, ax =plt.subplots()
ax.set_title('Survival During Treatment')
#get the data for each drug treatment (no error bars as these are not means, just straight counts)
ax.plot(timepoints, mice_count_expanded['Capomulin']*100/max(mice_count_expanded['Capomulin']),
        '--',color='r', marker='o', label='Capomulin')
ax.plot(timepoints, mice_count_expanded['Infubinol']*100/max(mice_count_expanded['Infubinol']),
        '--',color='b', marker='^', label='Infubinol')
ax.plot(timepoints, mice_count_expanded['Ketapril']*100/max(mice_count_expanded['Ketapril']),
        '--',color='g', marker='s', label='Ketapril')
ax.plot(timepoints, mice_count_expanded['Placebo']*100/max(mice_count_expanded['Placebo']),
        '--',color='black', marker='d', label='Placebo')
#format our plot
ax.set_xlim(min(timepoints)-2, max(timepoints)+2)
ax.set_ylim(min(min(mice_count_expanded['Capomulin']*100/max(mice_count_expanded['Capomulin'])),
               min(mice_count_expanded['Infubinol']*100/max(mice_count_expanded['Infubinol'])),
               min(mice_count_expanded['Ketapril']*100/max(mice_count_expanded['Ketapril'])),
               min(mice_count_expanded['Placebo']*100/max(mice_count_expanded['Placebo'])))-5,
           max(max(mice_count_expanded['Capomulin']*100/max(mice_count_expanded['Capomulin'])),
              max(mice_count_expanded['Infubinol']*100/max(mice_count_expanded['Infubinol'])),
              max(mice_count_expanded['Ketapril']*100/max(mice_count_expanded['Ketapril'])),
              max(mice_count_expanded['Placebo']*100/max(mice_count_expanded['Placebo'])))+5)
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Survival Rate (%)')
ax.legend(loc='best')
ax.grid()
# Save the Figure
plt.savefig('Images/survival.png')
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
volume_changes=mean_volume_data.unstack('Timepoint')
volume_changes.columns=volume_changes.columns.droplevel(0)
percent_changes=(volume_changes[45]-volume_changes[0])*100/volume_changes[0]
# Display the data to confirm
percent_changes
```

```python
# Store all Relevant Percent Changes into a Tuple
relevant_percents=(percent_changes.loc['Capomulin'],
                   percent_changes.loc['Infubinol'],
                   percent_changes.loc['Ketapril'],
                   percent_changes.loc['Placebo'])
Drugs=('Capomulin', 'Infubinol', 'Ketapril', 'Placebo')

# Create a color dictionary to splice the data between passing and failing drugs
colordict=[]
for i in np.arange(len(relevant_percents)):
    if relevant_percents[i]>0:
        colordict.append('red')
    else:
        colordict.append('green')
# Plot our data
plt.bar(x=Drugs,height=relevant_percents, color=colordict, align='center', width=1, edgecolor='black')
# Format our plot 
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")
plt.grid(axis='y')
plt.hlines(0, -0.75, len(Drugs)-0.25, color='black')
plt.xlim(-0.75, len(Drugs)-0.25)
plt.ylim(min(relevant_percents)-5, max(relevant_percents)+5)

# Use functions to label the percentages of changes
for i in np.arange(len(relevant_percents)):
    plt.annotate('{:,.2%}'.format(relevant_percents[i]/100),(i-0.25,5*relevant_percents[i]/abs(relevant_percents[i])))

# Save the Figure
plt.savefig('Images/change.png')

# Show the Figure
plt.show()
```
